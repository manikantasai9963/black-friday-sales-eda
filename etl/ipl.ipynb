```python
matches = matches.dropna(subset=['winner'])
matches['city'].fillna("Unknown", inplace=True)
```

```python
matches.replace('Delhi Daredevils','Delhi Capitals',inplace=True)
matches.replace('Kings XI Punjab','Punjab Kings',inplace=True)
deliveries.replace('Delhi Daredevils','Delhi Capitals',inplace=True)
deliveries.replace('Kings XI Punjab','Punjab Kings',inplace=True)
```

```python
df = deliveries.merge(matches,
                      left_on='match_id',
                      right_on='id')

df.head()
```

```python
df['total_runs_ball'] = df['total_runs']
```

```python
match_runs = df.groupby('match_id')['total_runs'].sum().reset_index()
match_runs.head()
```

```python
team_runs = df.groupby(['match_id','batting_team'])['total_runs'].sum().reset_index()
team_runs.head()
```

```python
team_total_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
team_total_runs
```

```python
team_matches = pd.concat([
    matches['team1'],
    matches['team2']
]).value_counts()

team_matches
```

```python
team_wins = matches['winner'].value_counts()
team_wins
```

```python
win_percentage = (team_wins / team_matches) * 100
win_percentage.sort_values(ascending=False)
```

```python
matches['winner'].value_counts()
```

```python
plt.figure(figsize=(10,6))
matches['winner'].value_counts().plot(kind='bar')

plt.title("Most Successful IPL Teams")
plt.xlabel("Team")
plt.ylabel("Wins")

plt.show()
```

```python
plt.figure(figsize=(10,6))
matches['season'].value_counts().sort_index().plot(kind='bar')

plt.title("Matches Per Season")
plt.show()
```

```python
sns.countplot(x='toss_decision', data=matches)

plt.title("Toss Decision Distribution")
plt.show()
```

```python
toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]

percentage = (len(toss_win_match_win)/len(matches))*100

print("Teams winning after toss:", percentage)
```

```python
team_total = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
team_total.plot(kind='bar')
plt.title("Total Runs by Team")
plt.show()
```

```python
wickets = df[df['is_wicket']==1]

top_bowlers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_bowlers.plot(kind='bar')
plt.title("Top 10 Wicket Takers")
plt.show()
```

```python
top_batsmen = df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_batsmen.plot(kind='bar')
plt.title("Top 10 Run Scorers")
plt.show()
```
